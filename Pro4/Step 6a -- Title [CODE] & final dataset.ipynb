```python
import pandas as pd
import numpy as np
import re, os

# Own config
import config

# Folder
folder = './training/Step5/'
dest_folder = './training/Final/'

dest_file =  dest_folder + 'dataset.csv'
```

```python
# Loading as datasets
entries =  os.listdir(folder)
print("Total {} files".format(len(entries)))

# Read file and push to list\n",
frame = pd.DataFrame()
# list_text.append([f.read() with open(file,'r') as f for file in entries"]

for entry in entries:
    file = folder + entry
    df_temp = pd.read_csv(file, sep='|', encoding="utf8")
    frame = pd.concat([frame, df_temp])
frame.head()
```

```python
frame.info()
```

```python
titles = frame['title'].values
exp = frame['experiences'].values
new_titles = [''] * len(titles)

for key,value in enumerate(titles):
#     print("value:",value)
    # CODE title
    code_titles = config.CODE_FIELDS
    temp_title = ''
    for k_title, v_title in code_titles.items():
#         print("[TITLE] with {} ------ {}".format(k_title, v_title))
        if value == k_title:
#             print("title mapping dict : ", k_title,' ------- ' ,value)
            temp_title = v_title
            break
    
    # CODE experiences
    code_exp = config.CODE_EXP
    temp_exp = ''
    for k_exp, v_exp in code_exp.items():
#         print("[EXP] with {} ------ {}".format(k_exp, v_exp))
        if exp[key] == k_exp:
#             print("exp mapping dict : ", k_exp,' ------- ' ,value)
            temp_exp = v_exp
            break
    
    # finalize CODE
    new_titles[key] = temp_exp + temp_title
#     print("================\n")
    
frame['title'] = new_titles
```

## Working with wage dataset

```python
frame_Wage = frame[frame['wage'] > 0]
frame_Wage.info()
```

```python
wage_dictionary = frame_Wage.groupby('title')['wage'].median()
# wage_dictionary = wage_dictionary.values().apply(list)
wage_dictionary = dict(wage_dictionary)
wage_dictionary
```

## Refill wage with median

```python
frame.info()
```

```python
frame['wage'].fillna(0,inplace=True)
```

```python
titles = frame['title'].values
wages = frame['wage'].values
new_wages = [''] * len(wages)

for key, value in enumerate(titles):
    print("B4: ", wages[key])
    for k_wag, v_wag in wage_dictionary.items():      
        
        print("------> k_wag: ",k_wag, value)        
        print("------> v_wag: ",v_wag)
        if (k_wag == value) and wages[key] == 0.0:
            print("Actual value: ", value)
            print("[WAGE] with {} ------ {}".format(k_wag, v_wag))
            print("While in [WAGE COLUMN]: ", wages[key], type(wages[key]))            
            print("wage mapping dict : ", k_wag,' ------- ' ,value)
            print("[FILL] with: ", v_wag)
                  
            new_wages[key] = v_wag
            break
        else:
            new_wages[key] = wages[key]
        print("===================\n")
            
            
frame['wage'] = new_wages
```
