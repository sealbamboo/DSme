```python
import pandas as pd
import numpy as np

file_url = './training/training_next_16052019.csv'
```

```python
frame = pd.read_csv(file_url, sep='|')
frame.head()
```

```python
frame.info()
```

```python
wtype = frame['type_empl'].values
new_wtype = [''] * len(wtype)

for key, value in enumerate(wtype):
    if value in 'Full Time' or value in 'Part Time':
        new_wtype[key] = 'Permanent'
    else:
        new_wtype[key] = 'Contract'

frame['type_empl'] = new_wtype
```

```python
frame['type_empl'].value_counts()
```

```python
frame['title'][100]
```

```python
titles = frame['title'].values
new_titles = [''] * len(titles)
title_leftover = [''] * len(titles)

for key, value in enumerate(titles):
    in_pos = value.find("Job")
    
    if in_pos:
        in_pos = in_pos - 1

        # Clean out "Job in ___  - SEEK" unused text
        new_titles[key] = value[0:in_pos].lstrip().rstrip()
        leftover = value[in_pos:]
        leftover = leftover.replace("Job in ","").replace(" - SEEK","")
        title_leftover[key] = leftover
        
    else:
        new_titles[key] = value
frame['title'] = new_titles
frame['mLocation'] = title_leftover
frame['mLocation'][0]
```
